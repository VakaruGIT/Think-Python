```python
from time import time
now = time()
now

seconds_in_day = 24 * 60 * 60
days_since_1970 = now // 86400
print("Days since 1970 =", int(days_since_1970), "days")

seconds_today = now % 86400

hours = seconds_today // 3600 + 1 # Austria
minutes = (seconds_today % 3600) // 60
seconds = seconds_today % 60

print(f"The time now is {int(hours)}H:{int(minutes)}MIN:{seconds:1f}S")
```

```python
def is_triangle(a, b, c):
    if a > b + c:
        print("No")
    elif b > a + c:
        print("No")
    elif c > a + b:
        print("No")
    elif a + b == c or b + c == a or a + c == b:
        print("We have a degenerate triangle")
    else:
        print("Yes")

is_triangle(4, 2, 2)
is_triangle(1, 3, 1)
is_triangle(2, 4, 5)
```

```python
from jupyturtle import forward, left, right, back, make_turtle

def draw(length):
    angle = 50
    factor = 0.6
    
    if length > 5:
        forward(length)
        left(angle)
        draw(factor * length)
        right(2 * angle)
        draw(factor * length)
        left(angle)
        back(length)

make_turtle(delay = 0)
draw(20)
```

```python
def koch(x):
    if x < 5:
        forward(x)
    else:
        koch(x/3)
        left(60)
        koch(x/3)
        right(120)
        koch(x/3)
        left(60)
        koch(x/3)

make_turtle(delay=0)
koch(120)

make_turtle(delay=0)
koch(5)
```

```python
def recurse(n,s):
    if n == 0:
        print(s)
    else:
        recurse(n-1, n+s)
        print(s) # Draw stack diagram to understand it

recurse(3, 0)
```
