## 6.4. Incremental development

```python
def distance(x1, y1, x2, y2):
    return 0.0

distance(1, 2, 4 ,6)
```

```python
def distance(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    print("dx is", dx)
    print("dy is", dy)
    return 0.0

distance(1, 2, 4, 6)
```

```python
def distance(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    dsquared = dx**2 + dy**2
    print("dsquared is: ", dsquared)
    return 0.0

distance(1, 2, 4, 6)
```

```python
import math

def distance(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    dsquared = dx**2 + dy**2
    result = math.sqrt(dsquared)
    print("result is", result)

distance(1, 2, 4, 6)
```

## 6.5. Boolean Functions

```python
def is_divisible(x, y):
    if x % y == 0:
        return True
    else:
        return False
```

```python
is_divisible(6,4)
```

```python
is_divisible(6, 3)
```

```python
# Function can be written more concisely
def is_divisible(x, y):
    return x % y == 0

if is_divisible(6, 2):
    print("divisible")
```
