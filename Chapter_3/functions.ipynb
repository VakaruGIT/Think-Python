# 3. Functions

## 3.1. Defining new functions

```python
def print_lyrics():
    print("I am a lumberjack, and I'm okay.")
    print("I sleep all night and I work all day.")

print_lyrics
print_lyrics()
```

## 3.2. Parameters

```python
def print_twice(string):
    print(string)
    print(string)

print_twice("Andrei Vacaru, ")

string = "Andrei Flavius, "
print(string)
print(string)

line = "Nono, "
print_twice(line)
```

## 3.3. Calling Function

```python
def repeat(word, n):
    print(word * n)

spam = "Spam, "
repeat(spam, 4)

def first_two_lines():
    repeat(spam, 4)
    repeat(spam, 4)

first_two_lines()

def last_three_lines():
    repeat(spam, 2)
    print("(Lovely Spam, Wonderful Spam!)")
    repeat(spam, 2)

last_three_lines()

def print_verse():
    first_two_lines()
    last_three_lines()

print_verse()
```

## 3.4. Repetition

```python
for i in range(2):
    print(i)

for i in range(2):
    print("Verse", i)
    print_verse()
    print()

def print_n_verses(n):
    for i in range(n):
        print_verse()
        print()
```

## 3.5. Variables and parameters are local

```python
def cat_twice(part1, part2):
    cat = part1 + part2
    print_twice(cat)

line1 = "Always look on the "
line2 = "bright side of life."
cat_twice(line1, line2)
```
