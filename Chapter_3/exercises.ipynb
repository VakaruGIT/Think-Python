```python
def print_right(string):
    length = len(string)
    print(" " * (40-length) + string)

print_right("Monty")
print_right("Python's")
print_right("Flying Circus")
```

```python
def triangle(string, integer):
    for i in range(integer + 1):
        print(string * i)

triangle("L", 5)
```

```python
def rectangle(string, length, height):
    for i in range(height):
        print(string * length)

rectangle("H", 5, 4)
    
```
