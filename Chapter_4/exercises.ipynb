```python
from jupyturtle import penup, pendown

def jump(length):
    """Move forward length units without leaving a trail

    Postcondition: leaves the pen down
    """
    penup()
    forward(length)
    pendown
```

## Rhombus

```python
def rhombus(length, angle = 60):
    parallelogram(length, length, angle)

make_turtle()
rhombus(50)
```

```python
def parallelogram(length1, length2, angle):
    forward(length1)
    left(angle)
    forward(length2)
    left(180 - angle)
    forward(length1)
    left(angle)
    forward(length2)
    left(180 - angle)

make_turtle()
parallelogram(50,50,60)
```

```python
def petal(radius, angle):
    arc(radius, angle)
    left(180 - angle)
    arc(radius, angle)

make_turtle(delay = 0)
petal(50, 50)

def flower(radius, angle, petals):
    for i in range(petals):
        petal(radius, angle)
        left(360 / petals)

make_turtle(delay = 0)
flower(30,80,11)
```
