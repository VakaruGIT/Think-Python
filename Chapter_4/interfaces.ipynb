## 4.3 Encapsulation and generalization

```python
def square():
    for i in range(4):
        forward(50)
        left(90)

make_turtle()
square()
```

```python
def square(length):
    for i in range(4):
        forward(length)
        left(90)

make_turtle()

square(30)
square(60)
```

```python
def polygon(n, length):
    angle = 360 / n
    for i in range(n):
        forward(length)
        left(angle)

make_turtle()
polygon(7, 30)
```

```python
import math 

def circle(radius):
    circumference = 2 * math.pi * radius
    n = 30
    length = circumference / n
    polygon(n, length)

make_turtle(delay = 0.01)
circle(30)
```

## 4.5. Refactoring

```python
def polyline(n, length, angle):
    for i in range(n):
        forward(length)
        left(angle)

def polyglon(n, length):
    angle = 360.0 / n
    polyline(n, length, angle)

def arc(radius, angle):
    arc_length = 2 * math.pi * radius * angle / 360
    n = 30
    length = arc_length / n
    step_angle = angle / n
    polyline(n, length, step_angle)

def circle(radius):
    arc(radius, 360)

make_turtle(delay=0.01)
polygon(n=20, length = 9)
arc(radius = 70, angle = 70)
circle(radius = 10)
```

## Development Plan
1. Start by writing a small program with no functions and definitions
2. Once you make it work, identify a coherent piece of it and encapsulate it in a function and give it a name
3. Generalize function by adding appropiate parameters
4. Repeat step 1 to 3 until you have a set of working function
5. Look for opportunities to improve the program by refactoring. One example could be repeated code in several places, factoring it to a general function is to take into consideration
