## 1.5. Values and types
- 2 is an integer value
- 2.0 is a floating-point value
- "2" is a string value
- '2' is also a string value
- True is a Boolean value
- False is also a Boolean value

```python
# Integers
print(type(42))  # <class 'int'>

# Floating-point numbers
print(type(3.14))  # <class 'float'>

# Strings
print(type("Hello"))  # <class 'str'>

# int, float, str, bool can be used to convert between types
print(int(3.99))  # 3 always rounds down
print(float(42))  # 42.0
print(str(42))    # "42"
print(bool(1))    # True
print(bool(0))    # False

# You can do tricky conversions
print(int("126") / 3)

# string into float
print(float("126.5"))
```

## 1.6. Formal and natural languages
- Formal languages are designed for specific applications

- Important differences between formal and natural languages:
  - Ambiguity: Natural languages are often ambiguous, while formal languages are designed to be unambiguous.
  - Redundancy: Natural languages often contain redundant information, while formal languages are designed to be concise.
  - Literalness: Formal languages are literal, while natural languages often use figurative language.

## 1.7. Debugging
- Debugging is the process of finding and fixing errors in a program.
- Common types of errors:
  - Syntax errors: Mistakes in the structure of the code that prevent it from being executed.
  - Runtime errors: Errors that occur while the program is running, such as division by zero.
  - Logical errors: Mistakes in the program's logic that produce incorrect results.

Strategies for debugging:
- Read error messages carefully to understand the problem.
- Use print statements to trace the flow of the program and identify where the error occurs.
- Break the program into smaller parts and test each part individually.
- Use a debugger tool to step through the code and inspect variables.

```python
# Bitwise operations
a = 60            # 60 = 0011 1100
b = 13            # 13 = 0000 1101

c = a & b        # AND operation: 12 = 0000 1100
print(f"Bitwise AND: {c}")
c = a | b        # OR operation: 61 = 0011 1101
print(f"Bitwise OR: {c}")
c = a ^ b        # XOR operation: 49 = 0011 0001
print(f"Bitwise XOR: {c}")
c = ~a           # NOT operation: -61 = 1100 0011
print(f"Bitwise NOT: {c}")
c = a << 2      # Left shift: 240 = 1111 0000
print(f"Bitwise Left Shift: {c}")
c = a >> 2      # Right shift: 15 = 0000 1111
print(f"Bitwise Right Shift: {c}")
```

```python
# Arithmetic questions
minutes = 42
seconds = 42
km = 10

# Time in seconds
total_seconds = (minutes * 60) + seconds
print(total_seconds)

# Distance in miles
miles = km / 1.61
print(miles)

# Seconds per mile
pace_seconds = total_seconds / miles
print(pace_seconds)

# Average pace (min and seconds)
pace_minutes = pace_seconds // 60
pace_remaining_seconds = pace_seconds % 60

print(int(pace_minutes), "minutes,", int(pace_remaining_seconds), "seconds per mile")

# Convert seconds to time
hours = total_seconds / 3600

avg_speed = miles / hours
print(avg_speed)
```
