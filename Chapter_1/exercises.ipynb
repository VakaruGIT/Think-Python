```python
# Bitwise operations
a = 60            # 60 = 0011 1100
b = 13            # 13 = 0000 1101

c = a & b        # AND operation: 12 = 0000 1100
print(f"Bitwise AND: {c}")
c = a | b        # OR operation: 61 = 0011 1101
print(f"Bitwise OR: {c}")
c = a ^ b        # XOR operation: 49 = 0011 0001
print(f"Bitwise XOR: {c}")
c = ~a           # NOT operation: -61 = 1100 0011
print(f"Bitwise NOT: {c}")
c = a << 2      # Left shift: 240 = 1111 0000
print(f"Bitwise Left Shift: {c}")
c = a >> 2      # Right shift: 15 = 0000 1111
print(f"Bitwise Right Shift: {c}")
```

```python
# Order of operations in python
result = 10 + 2 * 3 ** 2 / (4 - 1)
# Step by step
# 1. Parentheses 3
# 2. Exponentiation 9
# 3. Multiply 18
# 4. Divide 6
# 5. Add 16
print(result)
```

```python
# Arithmetic questions
minutes = 42
seconds = 42
km = 10

# Time in seconds
total_seconds = (minutes * 60) + seconds
print(total_seconds)

# Distance in miles
miles = km / 1.61
print(miles)

# Seconds per mile
pace_seconds = total_seconds / miles
print(pace_seconds)

# Average pace (min and seconds)
pace_minutes = pace_seconds // 60
pace_remaining_seconds = pace_seconds % 60

print(int(pace_minutes), "minutes,", int(pace_remaining_seconds), "seconds per mile")

# Convert seconds to time
hours = total_seconds / 3600

avg_speed = miles / hours
print(avg_speed)
```

```python
# Extra from LLM
tank = 500
fill_rate_ml_sec = 150

# Step 1: Convert base units
total_ml = tank * 1000
total_seconds = total_ml / fill_rate_ml_sec

# Step 2: Break down total seconds
minutes = int(total_seconds // 60)
remaining_seconds = total_seconds % 60

# Nice output
print(f"Tank file fills in {minutes} minutes and {remaining_seconds:.1f} seconds")
```
